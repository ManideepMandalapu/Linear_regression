```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#Loading data from local machine
data = pd.read_csv("house_dataset.csv")
```

```python
# Training the model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset
data = pd.read_csv("house_dataset.csv")

def normalize(feature):
    """Standardize the feature using Z-score normalization."""
    return (feature - np.mean(feature)) / np.std(feature)

# Normalize data
X = normalize(np.array(data['area']))
y_normalized = normalize(np.array(data['price']))

# Hyperparameters
alpha = 0.01
num_iterations = 10000

# Initialization
m = len(data['area'])
theta0 = 0  # intercept
theta1 = 0  # slope

# Gradient Descent
for _ in range(num_iterations):
    # Predictions
    y_pred = theta0 + theta1 * X
    
    # Derivatives
    dtheta0 = (1/m) * sum(y_pred - y_normalized)
    dtheta1 = (1/m) * sum((y_pred - y_normalized) * X)
    
    # Update thetas
    theta0 = theta0 - alpha * dtheta0
    theta1 = theta1 - alpha * dtheta1

# Visualization
plt.scatter(X, y_normalized, color='blue')
plt.plot(X, theta0 + theta1 * X, color='red')
plt.title('Normalized Housing Price vs Normalized Area')
plt.xlabel('Normalized Area')
plt.ylabel('Normalized Price')
plt.show()
```
